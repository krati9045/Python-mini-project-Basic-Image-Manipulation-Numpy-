# Basic Image Manipulation using Numpy

# 1. DOCUMENTATION

```python
This project includes an interactive program for basic image editing using Numpy. It is a menu driven program which accepts a choice(ch) from the user and performs the following operations: For ch=1, Load the Image

For ch=2, Edit the Image Edit Imageâ€™ includes the following operations which can be choosen by user(z): For z=1: Horizontal Flip For z=2: Vertical Flip For z=3: Invert Colors For z=4: Convert to Grayscale For z=5: Convert to B&W If user enters any other choice, it gives the alert message "Invalid choice".

For ch=3, Save the Image

For ch=4, Quit

If user enters any other choice, it gives the alert message "Invalid choice".
```

# 2. CODE

```python
#Importing the libraries
from skimage import io
from skimage import color
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import PIL.ImageOps 
```

```python
k=1
while k==1:

    print('*'*25)
    print("Press 1 to Load the image",end="\n")    #Menu for the program
    print("Press 2 to Edit the image",end="\n")
    print("Press 3 to Save the image",end="\n")
    print("Press 4 to Quit",end="\n")
    print('*'*25)
    
    pause_time = 0.2

    ch=int(input("Enter the choice"))

    if ch==1:                                 #Load image
        np_image=io.imread('image.jpeg') 
        np_array=np.array(np_image)
        print("Image is Loaded")


    elif ch==2:
        print('*'*25)                  #Edit the image
        print("Press 1 to Horizontal flip",end="\n")
        print("Press 2 to Vertical flip",end="\n")     #Menu for editing the image
        print("Press 3 to Invert colours",end="\n")
        print("Press 4 to Convert to grayscale",end="\n")
        print("Press 5 to Convert to B&W",end="\n")
        print('*'*25)

        z=int(input("Enter the choice"))

        if z==1: 
            np_image_hflip=np.fliplr(np_image) 
            plt.title('Horizontal Flip')    #Horizontal Flip
            plt.imshow(np_image_hflip)
            plt.show()


        elif z==2:
            np_image_vflip=np.flipud(np_image)    #Vertical Flip
            plt.title('Vertical Flip')
            plt.imshow(np_image_vflip)
            plt.show()

        elif z==3:                        #Invert Colours
            image = Image.open('image.jpeg')
            inverted_image = PIL.ImageOps.invert(image)
            plt.title('Invert Colours')
            plt.imshow(inverted_image)
            plt.show()
            
        elif z==4:
            img = io.imread('image.jpeg')     #Convert to Grayscale
            imgGray = color.rgb2gray(img)
            plt.title('Grayscale Image')
            plt.imshow(imgGray)
            plt.show()
            
        elif z==5:
            img = Image.open('image.jpeg')
            bw_img=img.convert("1")
            plt.title('Black and White Image')   #Convert to B&W
            plt.imshow(bw_img)
            plt.show()
            
        else:
            print("Invalid choice")

    elif ch==3:
        print("Image is saved")
        plt.imsave('image.jpeg',np_array)
        plt.imshow(np_image)         #Save the Image
        plt.show()

    elif ch==4:
        print("Quit")
        k=0           #Quit
        break

    else:
        print("Invalid choice")      #For invalid choice
    
    
```
